```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CSV dosyasından veri okunarak data'ya aktarılır
df = pd.read_csv('genel_arac_verileri_gorsellestirme_icin.csv')
```

```python
# 'marka' sütunundaki benzersiz değerlerin histogramını çiz
df['marka'].value_counts().plot(kind='bar', color='skyblue')

# Grafik başlığı ve ekseni etiketlerini ekleyin
plt.title('Marka Değerlerinin Histogramı')
plt.xlabel('Marka')
plt.ylabel('Frekans')

# Şemayı göster
plt.show()
```

```python
import csv
# CSV dosyasını oku
with open('genel_arac_verileri_gorsellestirme_icin.csv', 'r', newline='') as dosya:
    csv_okuyucu = csv.reader(dosya)
    satirlar = list(csv_okuyucu)

istanbul_sayisi = 0
ankara_sayisi = 0
izmir_sayisi = 0 
kocaeli_sayisi = 0
bursa_sayisi = 0

ilan_konum_indis = 1
for satir in satirlar:
    if(satir[ilan_konum_indis] == 'ilan_konumu'):
        continue
    elif(satir[ilan_konum_indis] == 'ï¿½stanbul'):
        istanbul_sayisi += 1
    elif(satir[ilan_konum_indis] == 'Ankara'):
        ankara_sayisi += 1
    elif(satir[ilan_konum_indis] == 'ï¿½zmir'):
        izmir_sayisi += 1
    elif(satir[ilan_konum_indis] == 'Kocaeli'):   
        kocaeli_sayisi += 1
    elif(satir[ilan_konum_indis] == 'Bursa'):   
        bursa_sayisi += 1

print("    ARAÇLARIN İLAN KONUMUNA GÖRE ŞEHİR SAYISI", "\n",
      "-------------------------------" ,"\n",
      "İstanbul sayisi: ", istanbul_sayisi, "\n",
      "Ankara sayisi: ", ankara_sayisi, "\n",
      "İzmir sayisi: ", izmir_sayisi, "\n",
      "Kocaeli sayisi: ", kocaeli_sayisi, "\n",
      "Bursa sayisi: ", bursa_sayisi, "\n",
      "-------------------------------" ,"\n")

ilan_sayilari = [istanbul_sayisi, ankara_sayisi, izmir_sayisi, kocaeli_sayisi, bursa_sayisi]
print(ilan_sayilari)
```

```python
ilan_sayilari = [istanbul_sayisi, ankara_sayisi, izmir_sayisi, kocaeli_sayisi, bursa_sayisi]
şehirler = ["İstanbul", "Ankara", "İzmir", "Kocaeli","Bursa"]

# Histogram grafiği oluştur
plt.figure(figsize=(10, 6))
sns.barplot(x=şehirler, y=ilan_sayilari, palette='viridis')

# Grafik başlığı ve ekseni etiketlerini ekleyin
plt.title('Şehirlere Göre İlan Sayısı')
plt.xlabel('Şehirler')
plt.ylabel('İlan Sayısı')

# Şemayı göster
plt.show()
```
